```python
import numpy as np
import matplotlib.pyplot as plt

# Define the slope and y-intercept
m = 2
c = 3

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values using the equation y = mx + c
y = m * x + c

# Plot the line
plt.plot(x, y, '-r', label='y = {}x + {}'.format(m, c))
plt.title('Linear Equation: y = {}x + {}'.format(m, c))
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
